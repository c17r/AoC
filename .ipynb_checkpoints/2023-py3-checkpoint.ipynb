```python
games = get_lines(2023, 2)

items = {
    "red": 12,
    "green": 13,
    "blue": 14,
}

def check_hands(hands):
    for color, max_num in items.items():
        for match in re.finditer(r"(\d+) " + color, hands):
            if int(match.group(1)) > max_num:
                return False
    return True

total = 0
for game in games:
    game_id, hands = game.split(":")
    if check_hands(hands):
        pieces = game_id.split(" ")
        total += int(pieces[-1])
print(total)

def min_colors(hands):
    rv = {}
    for color in items.keys():
        rv[color] = 0
        matches = list(re.finditer(r"(\d+) " + color, hands))
        if len(matches) > 0:
            t = [m.group(1) for m in matches]
            print(t)
            i = [int(a) for a in t]

    return rv

games = ['Game 1: 7 blue, 9 red, 1 green; 8 green; 10 green, 5 blue, 3 red; 11 blue, 5 red, 1 green']
for game in games:
    game_id, hands = game.split(":")
    print(min_colors(hands))
```

```python
@timeit
def day1():
    lines = get_lines(2023, 1)

    @timeit
    def part1():
        def parse(line):
            ints = re.findall(r"\d", line)
            return int(ints[0] + ints[-1])
            
        return sum([parse(line) for line in lines])
    handle_result(part1(), 2023, 1, "a")

    @timeit
    def part2():
        convert = { "one": "1", "two": "2", "three": "3", 
                    "four": "4", "five": "5", "six": "6", 
                    "seven": "7", "eight": "8", "nine": "9", 
                    r"\d": None }
        
        def get_digit(match):
            val = match.group(0)
            return convert.get(val, val)
        
        def parse(line):
            matches = []
            for check in convert.keys():
                matches.extend(list(re.finditer(check, line)))
            matches.sort(key=lambda m: m.start())
            digits = [get_digit(m) for m in matches]
            return int(digits[0] + digits[-1])
            
        return sum([parse(line) for line in lines])
    handle_result(part2(), 2023, 1, "b")

day1()
```
